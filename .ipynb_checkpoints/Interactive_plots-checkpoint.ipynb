```python
import matplotlib
import pandas as pd
from plotly import graph_objects as go
import plotly.offline as py 
import cufflinks as cf
import numpy as np


#py.offline.init_notebook_mode(connected=False)
cf.go_offline()
import plotly.io as pio
from IPython.display import IFrame
pio.renderers.default = 'iframe'
```

```python
df = pd.read_csv('data/clean_dataset.csv', delimiter = ',')
```

```python
df.head()
```

```python
df_plot = df[['risk', 'results', 'inspection_type']]
```

```python
risk = df_plot['risk'].value_counts()
results = df_plot['results'].value_counts()
inspect_type = df_plot['inspection_type'].value_counts()
```

```python
fig = go.Figure()


# set up the first trace
fig.add_trace(go.Bar(
                    x = results.index,
                    y = results,
                    visible = True
                )
             )
# set up the second trace
fig.add_trace(go.Bar(
                    x = risk.index,
                    y = risk,
                    visible=False
                )
             )

fig.add_trace(go.Bar(
                    x = inspect_type.index,
                    y = inspect_type,
                    visible = False
                )
             )

fig.layout.update(
    width=800,
    height=900,
    autosize=False,
    #margin=dict(t=0, b=0, l=0, r=0),
    template="plotly_white",
    yaxis=go.layout.YAxis(
        tickmode="array",
        titlefont=dict(size=30),
    )
)


annotations = [dict(
        text = "Food inspections dataset"
)]



fig.layout.update(
    updatemenus=[
        go.layout.Updatemenu(
            buttons=list([
                dict(
                    args=[{'visible': [True, False, False]}],
                    label="Results",
                    method="update"
                ),
                dict(
                    args=[{'visible': [False, True, False]}],
                    label="Risk",
                    method="update"
                ),
                dict(
                    args=[{'visible': [False, False, True]}],
                    label="Inspection type",
                    method="update"
                )
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)



# add dropdown menus to the figure
fig.layout.update(
    annotations=[
        go.layout.Annotation(text="Trace type:", showarrow=False,
                             x=0, y=3000, yref="paper", align="left")
    ]
)

fig.show()
```

```python
# for website plotting 
py.plot(fig, include_plotlyjs=False, output_type='div')
```
